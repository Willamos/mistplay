# Pytorch

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
from torch.utils.data.sampler import SubsetRandomSampler
```

# getting the data ready

```python
scaled_inputs = torch.from_numpy(scaled_inputs)
scaled_inputs = scaled_inputs.float()
labels = torch.from_numpy(np.array(labels))
# labels = labels.float()
```

```python
print(inputs.shape, labels.shape)
```

```python
class MyDataset(torch.utils.data.Dataset):
    def __init__(self, inputs, labels):
        self.inputs = inputs
        self.labels = labels
        
    def __len__(self):
        return len(self.inputs)
    
    def __getitem__(self, index):
        return self.inputs[index], self.labels[index]
```

```python
train_data = MyDataset(scaled_inputs, labels)
```

# getting a validation set

```python
valid_size = .2
num_train = len(train_data)
indices = list(range(num_train))
np.random.shuffle(indices)
split = int(np.floor(valid_size*num_train))

train_idx, valid_idx = indices[split:], indices[:split]

train_sampler = SubsetRandomSampler(train_idx)
valid_sampler = SubsetRandomSampler(valid_idx)
```

```python
train_loader = torch.utils.data.DataLoader(train_data, batch_size=64, 
                                           sampler=train_sampler, num_workers=10)
valid_loader = torch.utils.data.DataLoader(train_data, batch_size=64,
                                          sampler=valid_sampler, num_workers=10)
```

```python
inputs, labels = next(iter(train_loader))
print(labels)
print(inputs.shape, labels.shape)
```

# creating an MLP

```python
class MLPClassifier(nn.Module):
    def __init__(self, in_size, hidden_layers, out_size):
        super().__init__()
        self.hiddens = nn.ModuleList([nn.Linear(in_size, hidden_layers[0])])
        hiddens_size = zip(hidden_layers[:-1], hidden_layers[1:])
        self.hiddens.extend([nn.Linear(h1, h2) for h1, h2 in hiddens_size])
        self.out = nn.Linear(hidden_layers[-1], out_size)
        self.dropout = nn.Dropout(.2)


    def forward(self, x):
        # x through hidden layers
        for layer in self.hiddens:
            x = self.dropout(F.relu(layer(x)))

        x = F.log_softmax(self.out(x), dim=1)

        return x
```

```python
n_inputs = 37
n_out = 2
hiddens = [40, 30]

model = MLPClassifier(in_size=n_inputs, out_size=n_out, hidden_layers=hiddens)
```

# checking if gpu is available

```python
device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
device
```

```python
model = model.to(device)
epochs = 50
optimizer = optim.SGD(model.parameters(), lr=3e-3)
criterion = nn.NLLLoss()
model
```

```python
train_losses, valid_losses, validation_accuracy = [], [], []

lower_valid_loss = np.Inf

for epoch in range(epochs):
    #####################
    #  train the model  #
    #####################
    train_loss = 0.0
    valid_loss = 0.0
    valid_acc = 0.0
    model.train()  
    for inputs, labels in train_loader:
        optimizer.zero_grad()
        inputs = inputs.to(device)
        labels = labels.to(device)

        output = model(inputs)
        
        loss = criterion(output, labels)
        loss.backward()
        optimizer.step()

        train_loss += loss.item()

    #####################
    #  valid the model  #
    #####################
    else:
        model.eval()
        valid_loss = 0.0
        valid_acc = 0.0
        batch_size = len(valid_loader)
        with torch.no_grad():
            for inputs, labels in valid_loader:
                inputs = inputs.to(device)
                labels = labels.to(device)

                log_ps = model(inputs)
                valid_loss += criterion(log_ps, labels)
                ps = torch.exp(log_ps)
                # calculating accuracy
                _, top_classes = ps.topk(1, dim=1)
                equals = top_classes == labels.view(*top_classes.shape)
                valid_acc += torch.mean(equals.type(torch.FloatTensor))
    
        train_loss = train_loss/len(train_loader)
        train_losses.append(train_loss)

        valid_loss /= len(valid_loader)
        valid_losses.append(valid_loss)

        valid_acc /= len(valid_loader)
        validation_accuracy.append(valid_acc)

        print('Epoch: {}/{}'.format(epoch+1, epochs),
              'training loss: {:.3f}'.format(train_losses[-1]),
              'validation loss: {:.3f}'.format(valid_losses[-1]),
              'valid accuracy: {:.3f}'.format(validation_accuracy[-1]))
        if lower_valid_loss > valid_loss:
            print('Saving the model: validation loss decreased ({:.4f} -> {:.4f})'.format(lower_valid_loss, valid_loss))
            torch.save(model.state_dict(), 'model.pt')
            lower_valid_loss = valid_loss
                
```
